```python
import os, shutil
```

```python
root_dir = r'E:\Images\Pankaj\Memories\Pune Trip - July 2023\Pune Trip - August 2023'
destination_dir = os.path.join(root_dir, 'toDelete')
```

```python
for dirpath, _, filenames in os.walk(root_dir):
    for filename in filenames:
            if ' 2.' in filename.upper():
                if os.path.exists(os.path.join(dirpath, filename.replace(' 2.','.'))):
                    print(f"Removing File - {filename}")
                    file_path = os.path.join(dirpath, filename)
                    if not os.path.exists(destination_dir):
                        os.makedirs(destination_dir)

                    # Move the file to the category folder
                    shutil.move(file_path, os.path.join(destination_dir, filename))
```
